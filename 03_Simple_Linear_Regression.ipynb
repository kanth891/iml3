# Experiment 3: Simple Linear Regression
**Objective:** Implement a simple Linear Regression model to perform regression on sample data.

```python
# Import Required Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# Part A: Linear Regression using Normal Equation (from scratch)

np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Visualize the data
plt.figure(figsize=(7, 5))
plt.scatter(X, y, color='blue', alpha=0.5)
plt.title("Sample Data")
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

```python
# Apply Normal Equation: theta = (X^T X)^-1 X^T y
X_b = np.c_[np.ones((100, 1)), X]   # Add bias term (x0 = 1)
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
print("Theta (Intercept, Slope):", theta_best.ravel())
```

```python
# Predict and Plot Regression Line
X_test = np.array([[0], [2]])
X_test_b = np.c_[np.ones((2, 1)), X_test]
y_predict = X_test_b.dot(theta_best)

plt.figure(figsize=(7, 5))
plt.plot(X_test, y_predict, 'r-', linewidth=2, label='Regression Line')
plt.scatter(X, y, color='blue', alpha=0.5, label='Data Points')
plt.title("Linear Regression – Normal Equation")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
# Part B: Linear Regression using Scikit-learn on House Price Dataset

df = pd.read_csv('datasets/house-prices.csv')
print("Dataset Shape:", df.shape)
print(df.head())
```

```python
# Train the Model
X_house = df[['Area']]
y_house = df['Price']

model = LinearRegression()
model.fit(X_house, y_house)

print("Intercept:", model.intercept_)
print("Coefficient (Slope):", model.coef_[0])
```

```python
# Evaluate the Model
y_pred = model.predict(X_house)
mse = mean_squared_error(y_house, y_pred)
rmse = np.sqrt(mse)
print(f"MSE  : {mse:.2f}")
print(f"RMSE : {rmse:.2f}")
```

```python
# Visualize Regression Line on House Price Data
plt.figure(figsize=(7, 5))
plt.scatter(df['Area'], df['Price'], color='red', marker='+', label='Actual Data')
plt.plot(df['Area'], y_pred, color='blue', linewidth=2, label='Regression Line')
plt.title("House Price vs Area")
plt.xlabel("Area (sq ft)")
plt.ylabel("Price (Lakhs)")
plt.legend()
plt.show()

# Predict for a new area
new_area = 2400
predicted_price = model.predict([[new_area]])
print(f"Predicted price for {new_area} sq ft: {predicted_price[0]:.2f} Lakhs")
```
